Built in function adalah fungsi bawaan dari python. Artinya pengguna tidak perlu membuat fungsinya, dan cukup memanggil fungsi tersebut.

Contoh dari built in function adalah:

<h2>Soal 3: Method dan Function</h2>

Jawab Pertanyaan di bawah ini:
    
- Apa perbedaan method dan function?

Perbedaan antara method dengan function yaitu:

1. Method adalah fungsi yang dimiliki oleh suatu object. Contoh dari method adalah lower() dan upper() yang berfungsi mengubah huruf menjadi huruf kecil dan huruf kapital. Method biasanya tidak dapat digunakan untuk objek lain, misalnya upper() dan lower() yang hanya dapat digunakan oleh tipe data string. 

2. Function adalah kumpulan kode yang akan dijalankan(run) jika nama fungsi tersebut dipanggil. Function biasanya memerlukan input argumen. Function dapat berupa built-in(bawaan) maupun dapat dibuat sendiri. Contoh dari function adalah sum() yang menjumlahkan semua bilangan yang berada di dalam list.

<h2>Soal 5: Membuat Simple Function</h2>

Buatlah suatu fungsi yang menerima satu input argumen berbentuk list dan mempunyai elemen bertipe numeric semua, dimana fungsi tersebut berguna untuk menghitung rata2 dari kumpulan elemen list tersebut. namai fungsi tersebut 'mean_list'

```python
obj_list = [11.25, 18.0, 20.0, 10.75, 9.50]
def mean_list(inp_list):
    # isikan code
    mean = sum(inp_list) / len(inp_list)
    return mean
print(mean_list(obj_list))
```

Expected Output:

13.9

<h2>Soal 6: Membuat Function dengan Multiple arguments</h2>

Buatlah suatu fungsi untuk melakukan penggabungan antara dua list

```python
obj_list = [2, 4, 5, 6]
obj_penambah = [1, 2, 3]
def kali_list(list_pertama, list_kedua):
    # isikan kode
    return list_pertama + list_kedua
    
print(kali_list(obj_list, obj_penambah))
```

Expected Output:

[2, 4, 5, 6, 1, 2, 3]
