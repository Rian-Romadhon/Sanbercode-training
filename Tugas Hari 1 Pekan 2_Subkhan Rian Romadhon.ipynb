<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab:

1. a. Perbedaan antara list dengan series yaitu, sumbu pada series dapat diubah menjadi bukan numerik seperti a,b,c,d dan untuk memanggil data dari series dapat dilakukan dengan memanggil index/sumbu tersebut. Index pada list tidak dapat diubah. 

 b. Perbedaan antara series dengan dictionary yaitu, key/indeks pada series dapat diduplikat sedangkan pada dictionary tidak bisa. Selain itu, slicing dapat dilakukan pada series sedangkan pada dictionary tidak dapat dilakukan.



2. Perbedaan antara dataframe dengan series yaitu, dataframe merupakan objek dua dimensi yang terdiri dari index baris dan index kolom. Sedangkan series merupakan objek satu dimensi. Series tipe datanya harus seragam sedangkan dataframe dapat berbeda antar kolom. 

```python
# code here
x = pd.Series(['Apel','Jeruk','Mangga','Jambu','Anggur','Leci','Melon'], index=['a','b','c','d','e','f','g'])
print(x)
```
