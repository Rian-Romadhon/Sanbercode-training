Jawab:

1. Data latih/training adalah data yang akan dipelajari oleh suatu model. Data validasi adalah data yang digunakan untuk proses validasi model dan mencegah overfitting. Data test adalah data yang digunakan untuk menguji model.

2. Cara menilai suatu model adalah dengan melihat seberapa besar akurasi yang diperoleh dari prediksi sebuah data. Data yang digunakan tidak boleh data training, tetapi data test yang diperoleh dengan membuat data baru atau dengan membagi data menggunakan train test split dari library sklearn. Menilai performa model juga dapat dilakukan dengan menggunakan cross validation yaitu membagi data menjadi sebanyak K bagian kemudian dilakukan iterasi menghitung akurasi sebanyak K kali , lalu dirata-rata. Cara ini merupakan cara yang lebih baik daripada menguji akurasi menggunakan data test.

3. Confusion matrix adalah metode penilaian performa model yang lebih tepat dari matriks akurasi. Metode ini membagi data (y_true dan y_pred) ke dalam empat bagian, yaitu True Positive(TP), True Negative(TN), False Positive(FP), dan False Negative(FN). Berikut penjelasan keempat komponen tersebut:

  1. True Positive adalah data positif yang diprediksi positif/benar.
  2. True Negative adalah data negatif yang diprediksi negatif/benar.
  3. False Positive adalah data negatif yang diprediksi salah/positif.
  4. False Negative adalah data positif yang diprediksi salah/negatif.

  Confusion matriks sangat cocok diaplikasikan pada data dengan kondisi   jumlah data yang tidak seimbang, misalnya jumlah data penderita kanker sebanyak 10 orang sedangkan yang tidak menderita kanker sebanyak 990 orang. Tentu jika kita hanya menggunakan matriks akurasi 10 orang yang menderita kanker akan terabaikan.

4. Classification report adalah penjabaran dari confusion matrix yang meliputi precision, recall, dan f1score. Ketiga rumus tersebut menggunakan empat komponen confusion matrix yang disebutkan di atas.
  1. precision = TP/(TP + FP)
  2. recall = TP/(TP + FN)
  3. f1score = 2 * precision * recall/(precision + recall)

Kali ini kita akan menggunakan data untuk memprediksi kelangsungan hidup pasien yang telah mengalami operasi payudara. Dengan informasi yang dimiliki terkait pasien, kita akan membuat model untuk memprediksi apakah pasien akan bertahan hidup dalam waktu lebih dari 5 tahun atau tidak.

Lebih Lengkapnya kalian bisa membaca informasi tentang dataset di link berikut: https://raw.githubusercontent.com/jbrownlee/Datasets/master/haberman.names

Buat model Klasifikasi (Model/Algoritma Bebas) untuk memprediksi status pasien dengan ketentuan sebagai berikut:

1. Bagi kedua data ini menjadi data training dan data test dengan test_size=0.25.
3. Pelajar tentang metrics roc_auc_score kemudian buatlah model dan evaluasi dengan menggunakan teknik cross-validation dengan scoring 'roc_auc'. Baca https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html untuk menggunakan metric roc_auc saat cross-validation.
3. Berapa score rata2 dari model dengan teknik cross-validation tersebut?
4. Prediksi data test dengan model yang telah kalian buat!
5. Bagaimana hasil confusion matrix dari hasil prediksi tersebut?
6. Bagaimana classification report dari hasil prediksi tersebut?
5. Seberapa baik model anda dalam memprediksi seorang pasien mempunyai status positive?
6. Seberapa baik model anda dalam memprediksi seorang pasien mempunyai status negatif?

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/haberman.csv'
list_cols = ['Age', "Patient's Years", "N_positive_ax", "survival_status"]
df = pd.read_csv(url, names=list_cols)
```

```python
df.head()
```

```python
df['survival_status'].value_counts()
```

```python
# Code here
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import scale

x = df.drop('survival_status', axis=1)
xs = scale(x)
y = df['survival_status']

x_train, x_test, y_train, y_test = train_test_split(xs, y, test_size=0.25, stratify=y, random_state=21)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5, weights='distance')
model.fit(x_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
val_score = cross_val_score(model ,xs, y, scoring='roc_auc', cv=10)
val_score
```

```python
val_score.mean()
```

```python
y_predict = model.predict(x_test)

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predict, labels=[1,2])
```

```python
#classification report
from sklearn.metrics import classification_report
class_report = classification_report(y_test, y_predict, labels=[2,1])
print(class_report)
```

7. Menurut saya model saya kurang baik dalam memprediksi seorang pasien mempunyai status positive meninggal kurang dari 5 tahun, karena nilainya dibawah 0.5

8. Menurut saya model saya sangat baik dalam memprediksi seorang pasien mempunyai status negatif meninggal kurang dari 5 tahun, karena nilainya lebih dari 0.5

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/haberman.csv'
list_cols = ['Age', "Patient's Years", "N_positive_ax", "survival_status"]
df = pd.read_csv(url, names=list_cols)
```

```python
df.head()
```

```python
#Code here
#Bagi kedua data berikut ini menjadi data training dan data test dengan test_size=0.25.
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import scale

x = df.drop('survival_status', axis=1)
xs = scale(x)
y = df['survival_status']

x_train, x_test, y_train, y_test = train_test_split(xs, y, test_size=0.25, stratify=y, random_state=21)
```

```python
#Gunakan algoritma KNN sebagai model classifier.
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np

param = {'n_neighbors':np.arange(2,80), 'weights':['uniform', 'distance']}
model = KNeighborsClassifier()

#Gunakan fungsi GridSearchCV untuk hyperparameter tuning dan model selection
gscv = GridSearchCV(model,param_grid=param, scoring='roc_auc', cv=5)
gscv.fit(xs, y)
```

```python
#Kombinasi hyperparameter terbaik menurut saya
gscv.best_params_
```

```python
#Score validasi terbaik
gscv.best_score_
```

```python
#Melatih data training
model = KNeighborsClassifier(n_neighbors=27, weights='uniform')
model.fit(x_train,y_train)
```

```python
model.classes_
#1. meninggal lebih dari 5 tahun
#2. meninggal kurang dari 5 tahun
```

```python
model.predict_proba(x_test)[:,1]
```

```python
#roc auc score untuk data testing
from sklearn.metrics import roc_auc_score
y_predict = model.predict(x_test)
rac = roc_auc_score(y_test, y_predict)
rac
```

Model saya termasuk baik karena nilai roc_auc_score dan validation score nya diatas 0.5

# Decision Tree Classifier

```python
#Gunakan algoritma DTC sebagai model classifier.
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np

param = {'max_depth':np.arange(2,20), 'splitter':['best', 'random']}        
model_tree = DecisionTreeClassifier()

#Gunakan fungsi GridSearchCV untuk hyperparameter tuning dan model selection
gscv = GridSearchCV(model_tree,param_grid=param, scoring='roc_auc', cv=5)
gscv.fit(xs, y)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
#Melatih data training
model_tree = DecisionTreeClassifier(max_depth=3, splitter='random')
model_tree.fit(x_train,y_train)
```

```python
model_tree.predict_proba(x_test)[:,1]
```

```python
from sklearn.metrics import roc_auc_score
y_predict = model_tree.predict(x_test)
rac = roc_auc_score(y_test, y_predict)
rac
```

Dari kedua algoritma di atas, menurut saya keduanya memiliki selisih yang sedikit terhadap roc auc score maupun validation scorenya. Akan tetapi saya lebih memilih algoritma KNN karena lebih besar roc_auc_score nya dibandingkan dengan decision tree classifier
