1. Machine learning adalah cabang dari ilmu kecerdasan buatan yang mempelajari tentang bagaimana suatu komputer dapat belajar dari data untuk meningkatkan kecerdasannya.

2. Data feature adalah kumpulan data/variabel yang berperan dalam memprediksi suatu kejadian, sedangkan data target merupakan hasil dari prediksi tersebut.

3. Supervised learning memiliki data label/target sedangkan unsupervised learning tidak mempunyai data tersebut. Oleh karena itu output dari supervised learning lebih jelas jika dibandingkan dengan unsupervised learning

4. Jenis-jenis supervised learning:
  *   Classification: outputnya berupa class label, misalkan memprediksi buah dengan output apel, jeruk, pisang.
  *   Regression: output berupa nilai yang continous, misalkan memprediksi harga rumah, outputnya berupa rentang harga

5. Hyperparameter adalah parameter yang didefinisikan pertama kali pada saat pembuatan objek model machine learning, sedangkan parameter didefinisikan di akhir. Jika dilihat berdasarkan kedudukannya, hyperparameter adalah bagian kecil dari parameter. Hyperparameter tidak mempengaruhi model, hanya mengoptimisasi model sedangkan parameter mempengaruhi model. Contoh dari parameter yaitu koefisien dan intercept pada linear regression, sedangkan contoh hyperparameter adalah fit_intercept yang digunakan untuk mengoptimisasi model linear regression.

6. Langkah-langkah mengaplikasikan algoritma dalam machine learning:

  1.   Memilih model atau algoritma yang digunakan
  2.   Memilih hyperparameter saat membuat model
  3.   Memisahkan data feature dan data target (sklearn hanya menerima data input dalam bentuk matrik bukan vektor)
  4.   Memerintahkan model untuk mempelajari data dengan method fit()
  5.   Mengaplikasikan model ke dalam data test dengan method predict() untuk supervised learning atau method transform() dan predict untuk unsupervised learning

<h1>Soal 2: Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

```python
# Code here

# memilih model yang digunakan
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree


# memilih hyperparameter dalam mpembuatan model
model = DecisionTreeRegressor(ccp_alpha=0.0, criterion='mse', max_depth=4,
                      max_features=None, max_leaf_nodes=7,
                      min_impurity_decrease=0.0, min_impurity_split=None,
                      min_samples_leaf=5, min_samples_split=2,
                      min_weight_fraction_leaf=0.0, presort='deprecated',
                      random_state=None, splitter='best')

# memisahkan data feature dan data target
x = 20 * rng.rand(50)
x_train = x.reshape(-1,1)

y = x**2 + 2 * x + - 1 + rng.randn(50)
y_train = y

# memerintahkan model untuk mempelajari data dengan method fit
model.fit(x_train,y_train)
```

```python
# mengaplikasikan model ke dalam data test

x_test = x_new.reshape(-1,1)
y_test = model.predict(x_test)
y_predict = model.predict(x_train)
```

```python
#plot data

fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(x, y, label='Data Asli')
ax.plot(x_test, y_test, label='Prediksi Terhadap Data Baru', color='green')
ax.scatter(x_train, y_predict, c='y', label='Prediksi Terhadap Data Training')
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.legend(bbox_to_anchor=(1.1,1.2))
plt.show()
```
